```python
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
```

```python
# Rough data
np.random.seed(42)
X = np.concatenate([np.random.normal(loc=[0, 0], scale=[1, 1], size=(100, 2)),
                    np.random.normal(loc=[5, 5], scale=[1, 1], size=(100, 2))])
```

```python
# Initialize the GMM
n_components = 2
gmm = GaussianMixture(n_components=n_components, random_state=42)
gmm.fit(X)
```

```python
# Calculation of means of each Gaussian component
means = gmm.means_
print("Means of the Gaussian components:\n", means)

# Calculate the covariances of each Gaussian component
covariances = gmm.covariances_
print("\nCovariances of the Gaussian components:\n", covariances)
```

```python
# Get the weights of each Gaussian component
weights = gmm.weights_
print("\nWeights of the Gaussian components:\n", weights)
```

```python
# Predict cluster labels
labels = gmm.predict(X)

# Plot the data points colored by their assigned cluster
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', alpha=0.7)
plt.scatter(means[:, 0], means[:, 1], c='red', marker='X', s=200, label='Component Means')
plt.title('GMM Clustering with Component Means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
